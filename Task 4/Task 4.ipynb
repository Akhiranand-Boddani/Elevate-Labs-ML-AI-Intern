```python
X = data.drop(columns=['id', 'diagnosis'])

y = data['diagnosis']
```

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, stratify=y, random_state=42)

X_train.shape, X_test.shape
```

```python
from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()

X_train_std = scaler.fit_transform(X_train)
X_test_std = scaler.transform(X_test)
```

```python
from sklearn.linear_model import LogisticRegression

logreg = LogisticRegression(max_iter=2000, random_state=42)
logreg.fit(X_train_std, y_train)
```

```python
y_pred = logreg.predict(X_test_std)
```

```python
from sklearn.metrics import (
    classification_report,
    roc_auc_score,
    precision_score,
    recall_score,
    fbeta_score,
    accuracy_score,
    confusion_matrix,
    roc_curve
)

print("Accuracy :", accuracy_score(y_test, y_pred))
print("ROC-AUC Score :", roc_auc_score(y_test, y_pred))
print("Precision :", precision_score(y_test, y_pred))
print("Recall :", recall_score(y_test, y_pred))
print("F0.5 - Score :", fbeta_score(y_test, y_pred, beta=0.5))
print("F1 - Score :", fbeta_score(y_test, y_pred, beta=1.0))
print("F2 - Score :", fbeta_score(y_test, y_pred, beta=2.0))
```

```python
print("classification_report : \n", classification_report(y_test, y_pred))
```

```python
proba_test = logreg.predict_proba(X_test_std)[:, 1]
y_pred_05 = (proba_test >= 0.5).astype(int)
```

```python
cm = confusion_matrix(y_test, y_pred)
# Confusion matrix plot
plt.figure(figsize=(6,4))
plt.imshow(cm, cmap="Blues")
plt.title("Confusion Matrix")
plt.colorbar()
plt.xticks([0,1], ["Pred M", "Pred B"])
plt.yticks([0,1], ["True M", "True B"])
for i in range(2):
    for j in range(2):
        plt.text(j, i, cm[i, j], ha="center", va="center", color="black", fontsize=12)
plt.tight_layout()
plt.show()
```

```python
# ROC Curve
fpr, tpr, roc_thresh = roc_curve(y_test, proba_test)
plt.figure(figsize=(5,4))
plt.plot(fpr, tpr, label=f"ROC (AUC = {roc_auc_score(y_test, proba_test):.3f})")
plt.plot([0,1], [0,1], "--", color="gray")
plt.xlabel("False Positive Rate")
plt.ylabel("True Positive Rate")
plt.title("ROC Curve")
plt.legend()
plt.tight_layout()
plt.show()
```

```python
from sklearn.metrics import precision_recall_curve

# Precision-Recall curve (helps understand threshold trade-offs)
prec, rec, pr_thresh = precision_recall_curve(y_test, proba_test)
plt.figure(figsize=(5,4))
plt.plot(rec, prec)
plt.xlabel("Recall")
plt.ylabel("Precision")
plt.title("Precision-Recall Curve")
plt.tight_layout()
plt.show()
```

```python
# Threshold tuning demonstration:
# Example: pick a threshold targeting higher recall (e.g., recall >= 0.97)
candidate_thresholds = np.linspace(0.1, 0.9, 17)
best_t, best_prec, best_rec = 0.5, None, None
print("\nThreshold sweep (precision, recall):")
for t in candidate_thresholds:
    y_pred = (proba_test >= t).astype(int)
    p = precision_score(y_test, y_pred, zero_division=0)
    r = recall_score(y_test, y_pred, zero_division=0)
    print(f"t={t:.2f} -> precision={p:.3f}, recall={r:.3f}")
    # choose threshold by preference; here, maximize F1 as an example
    # but keeping it simple and not defining custom functions:
# Compute F1 inline for selection (without a function)
f1_scores = []
for t in candidate_thresholds:
    y_pred = (proba_test >= t).astype(int)
    p = precision_score(y_test, y_pred, zero_division=0)
    r = recall_score(y_test, y_pred, zero_division=0)
    if p + r == 0:
        f1 = 0.0
    else:
        f1 = 2 * p * r / (p + r)
    f1_scores.append(f1)
```

```python
best_idx = int(np.argmax(f1_scores))
best_threshold = float(candidate_thresholds[best_idx])
print(f"\nBest threshold by F1 approx: {best_threshold:.2f}, F1={f1_scores[best_idx]:.3f}")

# Evaluate at best threshold
y_pred_best = (proba_test >= best_threshold).astype(int)
cm_best = confusion_matrix(y_test, y_pred_best)
print("Confusion matrix @best threshold:\n", cm_best)
print("Precision @best:", precision_score(y_test, y_pred_best))
print("Recall @best:", recall_score(y_test, y_pred_best))
```
